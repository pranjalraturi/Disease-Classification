```python
import matplotlib.pyplot as m
import numpy as n
import pandas as p
noise=n.random.random(size=(100,1))
x_train=3* n.random.random(size=(100,1))
y_train=9 + 2*x_train +noise
m.scatter(x_train,y_train,color="royalblue")
m.xlabel("Brightness",color='m')
m.ylabel("Size",color='m')
m.title("Scatter plot of Brightness VS Size",weight='bold',color='maroon')
m.show()
```

```python
m.scatter(x_train,y_train-noise,color="royalblue")
m.xlabel("Brightness",color='m')
m.ylabel("Size",color='m')
m.title("Scatter plot of Brightness VS Size",weight='bold',color='maroon')
m.show()
```

```python
w=0
b=0
learning_rate=0.01
```

```python
def predict(x,w,b):
  return w*x+b

def cost_function(x,y,w,b):
  y_cap=predict(x,w,b)
  mse_cost=(y-y_cap)**2
  return n.mean(mse_cost)
```

```python
cost_function(x_train,y_train,w,b)
```

```python
def update_weights(x,y,w,b,learning_rate):

  dLdw= n.mean(-2*x*(y-predict(x,w,b)))
  dLdb= n.mean(-2*(y-predict(x,w,b)))

  w=w-learning_rate*dLdw
  b=b-learning_rate*dLdb

  return w,b
```

```python
update_weights(x_train,y_train,w,b,learning_rate)
```

```python
def train(x,y,w,b,learning_rate, tol=1e-13,n_epochs=100, verbose=False):
  weights=[w]
  biases = [b]
  costs = []

  ct=1
  while True:
    cost = cost_function(x,y,w,b)
    costs.append(cost)

    if len(costs) > 1 and abs(costs[-1] - costs[-2]) < tol :
      break

    w,b= update_weights(x,y,w,b,learning_rate)

    weights.append(w)
    biases.append(b)
    if verbose and ct %   n_epochs ==0:
      print("After ",ct," :- W= ",w," b= ",b," cost= ",cost)
    ct+=1
  return weights, biases , costs , ct

```

```python
weights,biases,costs,iteration_count=train(x_train,y_train,w,b,learning_rate,verbose=True)
```

```python
initial = [weights[0], biases[0], costs[0]]
final = [weights[-1], biases[-1], costs[-1]]
df=p.DataFrame(list(zip(initial,final)),
               columns=['Initial','Final'],
               index = ['Weight','Bias','cost'])
df
```

```python
m.figure(figsize=(10,4))
m.plot(costs,color='m')
m.xlabel('Epochs',color='blue')
m.ylabel('Costs',color='blue')
m.title('Costs Vs Epochs',weight='bold',color='maroon')
m.show()
```

```python
x_test=3*n.random.rand(10,1)
y_test=9+ 2*x_test + n.random.rand(10,1)

prediction=predict(x_test,weights[-1], biases[-1])

df1=p.DataFrame(list(zip(x_test.reshape(10,), y_test.reshape(10,), prediction.reshape(10,))),
                columns=['Test Input','True Input','Predicted Output'])
df1
```

```python
mse=n.mean((y_test-prediction)**2)
print(round(mse,3))
```
